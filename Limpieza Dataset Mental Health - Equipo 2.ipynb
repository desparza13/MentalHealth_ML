### Descripción de los features

- **Timestamp:** Fecha y hora en la que registró esa respuesta a la encuesta
- **Age:** Edad del encuestado
- **Gender:** Género del encuestado
- **Country:** País del encuestado
- **state:** Estado de Estados Unidos del encuestado, solo aplicable si pertenece a Estados Unidos
- **self_employed:** Pertenece a una empresa o trabaja por su cuenta
- **family_history:** Miembros de su familia cercana tienen una enfermedad mental
- **treatment:** Si ha buscado tratamiento para alguna condición de salud mental
- **work_interfere:** Si tiene una enfermedad mental si esta interfiere con su trabajo
- **no_employees:** Cantidad de empleados que tiene la compañia u organización
- **remote_work:** Si trabaja remotamente al menos el 50% del tiempo
- **tech_company:** La empresa para la que trabaja es principalmente una compañía tecnológica
- **benefits:** El empleador provee beneficios relacionados con la salud mental
- **care_options:** Conoce las opciones relacionados con el cuidado de la salud mental que el empleador provee
- **wellness_program:** El empleador ha discutido alguna vez la salud mental como parte del programa de bienestar para el empleado
- **seek_help:** El empleador provee recursos para aprender más sobre salud mental y cómo buscar ayuda
- **anonymity:** Se protege la identidad de la persona si toma la decisión de buscar tratamiento de alguna enfermedad mental
- **leave:** Que tan sencillo es obtener permiso de faltar por una condición de salud mental
- **mental_health_consequence:** Si cree que el discutir algún tema relacionado con la salud mental con su empleador ocasionaría consecuencias negativas
- **phys_health_consequence:** Si cree que el discutir algún tema relacionado con la salud física con su empleador ocasionaría consecuencias negativas
- **coworkers:** Si estaría dispuesto a discutir algún tema relacionado con la salud mental con sus compañeros de trabajo
- **supervisor:** Si estaría dispuesto a discutir algún tema relacionado con la salud mental con su supervisor directo
- **mental_health_interview:** Si estaría dispuesto a discutir su salud mental con un potencial empleador en una entrevista
- **phys_health_interview:** Si estaría dispuesto a discutir su salud física con un potencial empleador en una entrevista
- **mental_vs_physical:** Si considera que su empleador toma tan seriamente la salud mental como la salud física
- **obs_consequence:** Si alguna vez ha observado o escuchado de consecuencias negativas hacia compañeros de trabajo con condiciones de salud mental
- **comments:** Notas y comentarios adicionales

## Modificación de valores binarios

```python
#Cambiar valores ambiguos para que concuerden con las descripciones de los features
#Care options se define como si conoce las opciones relacionados con el cuidado de la salud mental que el empleador provee
#por ello si el encuestado contesta Not sure asumimos que es un no, no conoce
df['care_options'] = df['care_options'].apply(lambda x: 'No' if x == 'Not sure' else x)
```

```python
#Wellness program se define como El empleador ha discutido alguna vez la salud mental como parte del programa de bienestar para el empleado
#por lo que si no sabe o no lo recuerda, el empleador no ha tenido esa discusión con el
df['wellness_program'] = df['wellness_program'].apply(lambda x:'No' if x == "Don't know" else x)
```

```python
#Coworkers se define como Si estaría dispuesto a discutir algún tema relacionado con la salud mental con sus compañeros de trabajo
#pero para fines de este estudio y poder generalizar, en caso de estar dispuesto a hablar con uno, lo tomaremos como si estar dispuesto a hablar
df['coworkers'] = df['coworkers'].apply(lambda x:'Yes' if x == "Some of them" else x)
```

```python
#Supervisor se define como Si estaría dispuesto a discutir algún tema relacionado con la salud mental con su supervisor
#pero para fines de este estudio y poder generalizar, en caso de estar dispuesto a hablar con uno, lo tomaremos como si estar dispuesto a hablar
df['supervisor'] = df['supervisor'].apply(lambda x:'Yes' if x == "Some of them" else x)
```

```python
#Desplegar las columnas con valores binarios
#No considere las que tienen maybe y asi
df[['self_employed','family_history','treatment','remote_work', 'tech_company', 'care_options', 'wellness_program', 'coworkers', 'supervisor','obs_consequence']]
```

```python
#Reemplazar las columnas 'family_history', 'treatment', 'remote_work', 'tech_company' y 'obs_consequence' por enteros para poder visualizar los datos
df['self_employed'] = df['self_employed'].apply(lambda x: 1 if x == 'Yes' else 0)
df['family_history'] = df['family_history'].apply(lambda x: 1 if x == 'Yes' else 0)
df['treatment'] = df['treatment'].apply(lambda x: 1 if x == 'Yes' else 0)
df['remote_work'] = df['remote_work'].apply(lambda x: 1 if x == 'Yes' else 0)
df['tech_company'] = df['tech_company'].apply(lambda x: 1 if x == 'Yes' else 0)
df['care_options'] = df['care_options'].apply(lambda x: 1 if x == 'Yes' else 0)
df['wellness_program'] = df['wellness_program'].apply(lambda x: 1 if x == 'Yes' else 0)
df['coworkers'] = df['coworkers'].apply(lambda x: 1 if x == 'Yes' else 0)
df['supervisor'] = df['supervisor'].apply(lambda x: 1 if x == 'Yes' else 0)
df['obs_consequence'] = df['obs_consequence'].apply(lambda x: 1 if x == 'Yes' else 0)

#Reemplazar los géneros por Hombre (0) y Mujer (1)
df['Gender'] = df['Gender'].apply(lambda x : 1 if x[0].lower() == 'f' else 0)

#Ver el cambio
df[['Gender', 'self_employed', 'family_history','treatment','remote_work', 'tech_company', 'obs_consequence']]
```

```python
#Reemplazar los valores de no_employees
df['no_employees'] = df['no_employees'].apply(lambda x: 0 if x == '1-5' else x)
df['no_employees'] = df['no_employees'].apply(lambda x: 1 if x == '6-25' else x)
df['no_employees'] = df['no_employees'].apply(lambda x: 2 if x == '26-100' else x)
df['no_employees'] = df['no_employees'].apply(lambda x: 3 if x == '100-500' else x)
df['no_employees'] = df['no_employees'].apply(lambda x: 4 if x == '500-1000' else 5)

#Reemplazar los valores de benefits
df['benefits'] = df['benefits'].apply(lambda x: 0 if x == 'No' else x)
df['benefits'] = df['benefits'].apply(lambda x: 1 if x == 'Yes' else 2)

#Reemplazar los valores de seek_help
df['seek_help'] = df['seek_help'].apply(lambda x: 0 if x == 'No' else x)
df['seek_help'] = df['seek_help'].apply(lambda x: 1 if x == 'Yes' else 2)

#Reemplazar los valores de anonymity
df['anonymity'] = df['anonymity'].apply(lambda x: 0 if x == 'No' else x)
df['anonymity'] = df['anonymity'].apply(lambda x: 1 if x == 'Yes' else 2)

#Reemplazar los valores de leave
df['leave'] = df['leave'].apply(lambda x: 0 if x == "Don't know" else x)
df['leave'] = df['leave'].apply(lambda x: 1 if x == 'Very easy' else x)
df['leave'] = df['leave'].apply(lambda x: 2 if x == 'Somewhat easy' else x)
df['leave'] = df['leave'].apply(lambda x: 3 if x == 'Somewhat difficult' else 4)

#Reemplazar los valores de mental_health_consequence
df['mental_health_consequence'] = df['mental_health_consequence'].apply(lambda x: 0 if x == 'No' else x)
df['mental_health_consequence'] = df['mental_health_consequence'].apply(lambda x: 1 if x == 'Yes' else 2)

#Reemplazar los valores de phys_health_consequence
df['phys_health_consequence'] = df['phys_health_consequence'].apply(lambda x: 0 if x == 'No' else x)
df['phys_health_consequence'] = df['phys_health_consequence'].apply(lambda x: 1 if x == 'Yes' else 2)

#Reemplazar los valores de mental_health_interview
df['mental_health_interview'] = df['mental_health_interview'].apply(lambda x: 0 if x == 'No' else x)
df['mental_health_interview'] = df['mental_health_interview'].apply(lambda x: 1 if x == 'Yes' else 2)

#Reemplazar los valores de phys_health_interview
df['phys_health_interview'] = df['phys_health_interview'].apply(lambda x: 0 if x == 'No' else x)
df['phys_health_interview'] = df['phys_health_interview'].apply(lambda x: 1 if x == 'Yes' else 2)

#Reemplazar los valores de mental_vs_physical
df['mental_vs_physical'] = df['mental_vs_physical'].apply(lambda x: 0 if x == 'No' else x)
df['mental_vs_physical'] = df['mental_vs_physical'].apply(lambda x: 1 if x == 'Yes' else 2)
```
